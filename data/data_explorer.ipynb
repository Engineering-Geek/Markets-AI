```python
df2 = pd.read_csv("data/raw/forex/CADJPY_M1.csv", delimiter="\t")[["Open", "High", "Low", "Close"]]
df2.head()
```

```python
frames = 20
std_dev = 2

df2['MA{}'.format(frames)] = df2['Close'].rolling(window=20).mean()
df2['{}STD'.format(frames)] = df2['Close'].rolling(window=20).std() 

df2['Upper'] = df2['MA{}'.format(frames)] + (df2['{}STD'.format(frames)] * std_dev)
df2['Lower'] = df2['MA{}'.format(frames)] - (df2['{}STD'.format(frames)] * std_dev)
```
